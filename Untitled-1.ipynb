```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
gt_path = '/nfs/turbo/jjparkcv-turbo-large/zichen/video-diffusion-pytorch/samples/obs=5000_obsw=10000/gt_seed0.pt'

dps_path = '/nfs/turbo/jjparkcv-turbo-large/zichen/video-diffusion-pytorch/samples/obs=5000_obsw=10000/dps.pt'

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print('Using {} device'.format(device))
```

```python

frame = 9

gt = torch.load(gt_path, map_location=device)
gt = gt[0][frame].numpy()

dps = torch.load(dps_path, map_location=device)
dps = dps[0][0][frame].numpy()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(gt)
ax[1].imshow(dps)
```

```python
relative_error = np.linalg.norm(gt - dps) / np.linalg.norm(gt)
print(f'Relative error: {relative_error}')

gt_mean = np.mean(gt)
rse = np.sum((gt - dps) ** 2) / np.sum(gt - np.mean(gt) ** 2) 
print(f'RSE: {rse}')
```
